### Extracting daily and hourly stock market data for Apple, October 2022 - October 2023 year

```python
import yfinance as yf
import pandas as pd

# Defining the ticker symbol for AAPL
ticker_symbol = "AAPL"

# The start and end dates for the data
start_date = "2022-10-01"
end_date = "2023-10-01"

# Downloading daily data
daily_data = yf.download(ticker_symbol, start=start_date, end=end_date)

# Download hourly data
hourly_data = yf.download(ticker_symbol, start=start_date, end=end_date, interval="1h")

# Print the first few rows of each data frame
print("Daily Data:")
print(daily_data.head())

print("\nHourly Data:")
print(hourly_data.head())
```

#### Creating plots in various ways

```python
#importing additional libraries

import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
daily_data.head()
```

```python
#Adding date as a new column

daily_data.reset_index(inplace=True)
daily_data.head()
```

```python
#Adding time as a new column

hourly_data.reset_index(inplace=True)
hourly_data.head()
```

```python
#renaming index

hourly_data.rename(columns={'index': 'Hour'}, inplace=True, errors='raise')
hourly_data.head()
```

#### Plotting the "Close" price by date

```python
plt.figure(figsize=(14,5))
sns.set_style("ticks")
sns.lineplot(data=daily_data,x="Date",y='Close',color='firebrick')
sns.despine()
plt.title("The Stock Price of Apple",size='x-large',color='blue')
```

#### Plotting the "Close" price by hour

```python
plt.figure(figsize=(14,5))
sns.set_style("ticks")
sns.lineplot(data=hourly_data,x="Hour",y='Close',color='firebrick')
sns.despine()
plt.title("The Stock Price of Apple by hour",size='x-large',color='blue')
```

### Create an overplot of the Simple Moving Average (SMA) of a 20-period and 50-period on a price

```python
df = daily_data
```

```python
df['SMA20'] = df['Close'].rolling(window=20).mean()
df['SMA50'] = df['Close'].rolling(window=50).mean()

#Pandas df.rolling() function provides the feature of rolling window calculations. 
#The concept of rolling window calculation is most primarily used in signal processing and time-series data. 
```

```python
# Create a new figure
plt.figure(figsize=(12, 6))

# Plot the price data
plt.plot(df['Date'], df['Close'], label='Price', color='blue')

# Plot SMA20 and SMA50
plt.plot(df['Date'], df['SMA20'], label='SMA20', color='red')
plt.plot(df['Date'], df['SMA50'], label='SMA50', color='green')

# Add labels and legend
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('The APPL Stock Price Chart with SMA20 and SMA50')
plt.legend()

# Show the plot
plt.show()
```
