### Extracting daily and hourly stock market data for Apple, October 2022 - October 2023 year

```python
import yfinance as yf
import pandas as pd

# Defining the ticker symbol for AAPL
ticker_symbol = "AAPL"

# The start and end dates for the data
start_date = "2022-10-01"
end_date = "2023-10-01"

# Downloading daily data
daily_data = yf.download(ticker_symbol, start=start_date, end=end_date)

# Download hourly data
hourly_data = yf.download(ticker_symbol, start=start_date, end=end_date, interval="1h")

# Print the first few rows of each data frame
print("Daily Data:")
print(daily_data.head())

print("\nHourly Data:")
print(hourly_data.head())
```

#### Creating plots in various ways

```python
#importing additional libraries

import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
daily_data.head()
```

**Open**: The opening price of the stock
    
**High**: The highest price of the stock
    
**Low**: The lowest price of the stock
    
**Close**: The closing price of the stock
    
**Adj Close**: Adjusted Closing Price of the stock
    
**Volume**: The total number of shares traded in a single day

```python
#Adding date as a new column

daily_data.reset_index(inplace=True)
daily_data.head()
```

```python
#Adding time as a new column

hourly_data.reset_index(inplace=True)
hourly_data.head()
```

```python
#renaming index

hourly_data.rename(columns={'index': 'Hour'}, inplace=True, errors='raise')
hourly_data.head()
```

#### Plotting the "Close" price by date

```python
plt.figure(figsize=(14,5))
sns.set_style("ticks")
sns.lineplot(data=daily_data,x="Date",y='Close',color='firebrick')
sns.despine()
plt.title("The Stock Price of Apple",size='x-large',color='blue')
```

#### Plotting the "Close" price by hour

```python
plt.figure(figsize=(14,5))
sns.set_style("ticks")
sns.lineplot(data=hourly_data,x="Hour",y='Close',color='firebrick')
sns.despine()
plt.title("The Stock Price of Apple by hour",size='x-large',color='blue')
```
